```python
# Importing the important libraries
import mlxtend
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

               ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

               ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

               ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

               ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]

te = TransactionEncoder()
te_ary = te.fit(toy_dataset).transform(toy_dataset)    #Apply one-hot-encoding on our dataset
data = pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
data
```

```python
from mlxtend.frequent_patterns import apriori
apriori(data, min_support=0.6)
```

```python
frequent_itemsets=apriori(data, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate items
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```

```python
cust_data = pd.read_csv('/content/Market_Basket_Optimisation.csv', header=None)
print(cust_data.head())
```

```python
transactions = cust_data.applymap(str).values.tolist()
transactions = [[item for item in transaction if item != 'nan'] for transaction in transactions]
transactions [:3]
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)   #Apply one-hot-encoding on our dataset
t_cust_data = pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
print(t_cust_data.head())
```

```python
apriori(t_cust_data, min_support=0.1)
```

```python
frequent_itemsets = apriori(t_cust_data, min_support=0.03, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.1) # associate items
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
