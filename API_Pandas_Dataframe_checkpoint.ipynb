```python
from types import ModuleType
import requests
from IPython.display import Image
import pandas as pd
```

```python
# Store API key in a variable
api_key = 'otfU6InuuH34NTXQBJkrhoXf9NjkrHJMfDE8qbpD'

# Making a request to the NASA APOD (Astronomy Picture of the Day) API
url = 'https://api.nasa.gov/planetary/apod'
params = {
    'api_key': api_key
}

data = requests.get(url, params=params)

# Check if the request was successful
# 200 == The requested action was successful.
if data.status_code == 200:
    data = data.json()
    print(data)
else:
    print(f"Error: {data.status_code}")
```

```python
# Extract the image URL from the API response
image_url = data['url']
image_title = data['title']

# Display the image
print(f"Title: {image_title}")
display(Image(url=image_url))
```

```python
# Make a request to the NeoWs API
url = f'https://api.nasa.gov/neo/rest/v1/feed?start_date=2020-09-07&end_date=2020-09-08&api_key={api_key}'

data = requests.get(url)

# Check if the request was successful
if data.status_code == 200:
    a_data = data.json()
    print(a_data)
else:
    print(f"Error: {data.status_code}")
```

```python
# Initialize lists to store the extracted data
asteroid_ids = []
asteroid_names = []
min_diameters_km = []
absolute_magnitudes = []
relative_velocities_kms = []

# Extract the data from the JSON response
for date in a_data['near_earth_objects']:
    for asteroid in a_data['near_earth_objects'][date]:
        asteroid_ids.append(asteroid['id'])
        asteroid_names.append(asteroid['name'])
        min_diameters_km.append(asteroid['estimated_diameter']['kilometers']['estimated_diameter_min'])
        absolute_magnitudes.append(asteroid['absolute_magnitude_h'])
        relative_velocities_kms.append(asteroid['close_approach_data'][0]['relative_velocity']['kilometers_per_second'])

# Create a DataFrame from the extracted data
asteriod_dataset = pd.DataFrame({
    'Asteroid ID': asteroid_ids,
    'Asteroid Name': asteroid_names,
    'Minimal Estimated Diameter (km)': min_diameters_km,
    'Absolute Magnitude': absolute_magnitudes,
    'Relative Velocity (km/s)': relative_velocities_kms
})

# Display the DataFrame
asteriod_dataset
```

```python
 # Export the DataFrame to a CSV file
asteriod_dataset.to_csv('asteroids_data.csv', index=False)
print("Data exported to 'asteroids_data.csv'")
```
